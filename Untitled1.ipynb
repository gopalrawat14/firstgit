```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://www.kaggle.com/c/titanic/data"
# Assuming you have downloaded the dataset locally as 'train.csv'
df = pd.read_csv('train.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())

# Check the summary statistics of numerical columns
print(df.describe())

# Check the data types and missing values
print(df.info())
```

```python
# Handling missing values
# Check missing values
print(df.isnull().sum())

# Impute missing values for 'Age' and 'Fare' with median, and 'Embarked' with mode
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Drop unnecessary columns like 'PassengerId', 'Name', 'Ticket', 'Cabin'
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Convert categorical variables to numeric using one-hot encoding
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Verify if there are any missing values left
print(df.isnull().sum())
```

```python
# Relationship between survival and other variables
sns.countplot(x='Survived', data=df)
plt.show()

sns.countplot(x='Survived', hue='Sex_male', data=df)
plt.show()

sns.countplot(x='Survived', hue='Pclass', data=df)
plt.show()

# Distribution of age among passengers
sns.histplot(df['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.show()

# Fare distribution
sns.histplot(df['Fare'], bins=20, kde=True)
plt.xlabel('Fare')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
